```python
from ask_delphi_api.authentication import AskDelphiClient
client = AskDelphiClient()
client.authenticate()   # pakt automatisch portal code uit .env
```

### Let op:
Onderstaande code werkt momenteel alleen voor zandbak 0

```python
from ask_delphi_api.topictools import TopicTools
from datetime import datetime
from ask_delphi_api.project import Project

project = Project(client)
topics = TopicTools(client)

# Maak een topic
# digitale coach procespagina
current_date = datetime.now()
topicId = topics.topic_upload("Nieuwe procespagina", f"Digitale Coach Procespagina - {current_date}")

# Haal topic-parts op.
content = topics.get_topic_parts(topicId=topicId)

# Selecteer part uit topic met daarin de content.
part = content['topicEditorData']['groups'][1]['parts'][0]

# Checkout topic.
result = topics.checkout(topicId)

# Selecteer huidig topicVersionId.
topicVersionId = result['topicVersionId']

# Pas content topic aan.
current_date = datetime.now()
topics.topic_add_content(topicVersionId=topicVersionId, topicId=topicId, partId="body", part=part, new_text=f'<p>Nieuwe datum: {current_date}</p>')

# Checkin topic.
topics.checkin(topicId)
```
