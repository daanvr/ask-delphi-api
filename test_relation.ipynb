```python
import os
import dotenv
import uuid

dotenv.load_dotenv()

os.getenv("ASKDELPHI_PORTAL_CODE")
```

```python
from askdelphi_client import AskDelphiClient

client = AskDelphiClient()

client.authenticate()
```

===== Get content design (topic types)

```python
try:
    design = client.get_content_design()

    topic_types = design.get("topicTypes", [])
    print(f"Found {len(topic_types)} topic types:\n")

    for i, tt in enumerate(topic_types[:100]):  # Show first 100
        print(f"  {i+1}. {tt.get('title', 'Unknown')}")
        print(f"     ID: {tt.get('key', 'N/A')}")
        print(f"     Namespace: {tt.get('namespace', 'N/A')}")
        print()

    if len(topic_types) > 100:
        print(f"  ... and {len(topic_types) - 100} more")

except Exception as e:
    print(f"FAILED: {e}")
```

===== Creates a new task topic with the specified title and type in the given project and tenant.

- <param name="context">API context containing authentication, tenant, project, and ACL information.</param>
- <param name="topicTitle">The title of the topic to create.</param>
- <param name="topicTypeId">The unique identifier of the topic type.</param>
- <returns>The identifier for the newly created topic version.</returns>

```python
# Create a task Topic
topic_id = str(uuid.uuid4())

result = client._request(
    "POST",
    "v4/tenant/{tenantId}/project/{projectId}/acl/{aclEntryId}/topic",
    json_data={
        "topicId": topic_id,
        "topicTitle": "API Test Topic f7cf7f58",
        "topicTypeId": "6aba8437-c8df-42d2-a868-840847c124ca",
         "copyParentTags": False
        }
)

topicId = result.get('topicId')
topicVersionKey = result.get('topicVersionKey')

print("Task topic created!")
print(f"  - Topic ID: {topicId}")
print(f"  - Version ID: {topicVersionKey}")
```

```python
endpoint = f"v1/tenant/{{tenantId}}/project/{{projectId}}/acl/{{aclEntryId}}/topic/{topic_id}/allowedrelations"

result = client._request(
    "GET",
    endpoint
)

allowedRelations = result.get("topicAllowedRelations", [])
print(f"Found {len(allowedRelations)} allowed topic relations:\n")

for i, tt in enumerate(allowedRelations[:100]):  # Show first 100
        print(f"  {i+1}. pyramidLevelName: {tt.get('pyramidLevelName', 'Unknown')}")
        print(f"     relationTypeId: {tt.get('relationTypeId', 'N/A')}")
        print(f"     topicTypeName: {tt.get('topicTypeName', 'N/A')}")
       
        print()

if len(allowedRelations) > 100:
    print(f"  ... and {len(allowedRelations) - 100} more")


print(f"Found {allowedRelations} \n")
```

Adds a relation to topic. Creates a relation from one topic version to another, using the specified relation type.

- <param name="context">API context containing authentication, tenant, project, and ACL information.</param>
- <param name="sourceTopicId">The identifier of the source topic version.</param>
- <param name="targetTopicId">The identifier of the target topic version.</param>
- <param name="relationType">The type of relation to create between the topics.</param>

```python
# Create a relation.
topic_id_1 = str(uuid.uuid4())
topic_id_2 = str(uuid.uuid4())

# Digitale Coach Procespagina
topic_type_id_1 = "7d332fbb-44f5-469f-b570-874e701e526b"
# Task
topic_type_id_2 = "6aba8437-c8df-42d2-a868-840847c124ca"

# Maak twee topics
result_1 = client._request(
    "POST",
    "v4/tenant/{tenantId}/project/{projectId}/acl/{aclEntryId}/topic",
    json_data={
        "topicId": topic_id_1,
        "topicTitle": "Test Topic 1",
        "topicTypeId": topic_type_id_1,
         "copyParentTags": False
        }
)

print("Task topic 1 created!")
print(f"  - Topic ID: {topicId}")
print(f"  - Version ID: {topicVersionKey}")

result_2 = client._request(
    "POST",
    "v4/tenant/{tenantId}/project/{projectId}/acl/{aclEntryId}/topic",
    json_data={
        "topicId": topic_id_2,
        "topicTitle": "Test Topic 2",
        "topicTypeId": topic_type_id_2,
         "copyParentTags": False
        }
)

print("Task topic 2 created!")
print(f"  - Topic ID: {topicId}")
print(f"  - Version ID: {topicVersionKey}")

# Checkout eerste topic voor bewerking.
endpoint = f"/v1/tenant/{{tenantId}}/project/{{projectId}}/acl/{{aclEntryId}}/topic/{topic_id_1}/workflowstate"
result = client._request("POST", endpoint,
                json_data={
                    "action": 1
                })
topicVersionId = result["topicVersionId"]

print(f"  - TopicVersion ID: {topicVersionId}")

# Voeg een relatie toe van dit topic naar andere topics.
# print(f" ADD Relation: Relatie toegevoegd van {topic_id_1} naar {topic_id_2}")

# Voeg een relatie toe van dit topic naar andere topics.
endpoint = f"v2/tenant/{{tenantId}}/project/{{projectId}}/acl/{{aclEntryId}}/topic/{topic_id_1}/topicVersion/{topicVersionId}/relation"
result = client._request(
    "POST", 
    endpoint,
    json_data={
        "relationTypeId": ["521ed333-5226-46f9-8765-1a2cea40fd37"],
        "sourceTopicId": topic_id_1,
        "targetTopicIds": [topic_id_2]
    })
```
