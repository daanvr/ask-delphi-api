```python
from askdelphi.authentication import AskDelphiClient
client = AskDelphiClient()
client.authenticate()   # pakt automatisch portal code uit .env
```

```python
from askdelphi.topictools import TopicTools
from askdelphi.project import Project

project = Project(client)

topics = TopicTools(client, project)
topicId = "dac77c17-b83c-4012-8da1-4a0f156639bc"
topic_version_id = topics.get_topicVersionId(topicId)
print(f"Version id topic before checkout : {topic_version_id}")

topics.checkout(topicId)

topic_version_id = topics.get_topicVersionId(topicId)
print(f"Version id topic after checkout : {topic_version_id}")
```

```python
topics.checkin(topicId)
topic_version_id = topics.get_topicVersionId(topicId)
print(f"Version id topic after checkin : {topic_version_id}")
```
