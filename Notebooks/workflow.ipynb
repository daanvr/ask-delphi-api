### Connectie met AskDelphi opzetten

```python
from askdelphi.authentication import AskDelphiClient
client = AskDelphiClient()
client.authenticate()   # pakt automatisch portal code uit .env
```

```python
from askdelphi.workflow import Workflow
workflow = Workflow(client)
```

### Creates a workflow transition request for specifiek topic.

```python
#  Creates a workflow transition request for specifiek topic.
topic_id_digicoach = "5defb4a3-d5f7-4a3e-9eac-598e25fb5f61"
request_id = workflow.create_workflow_transition_request(topic_id_digicoach)
print(f"Created workflow transition request ID : {request_id}")

# Gets workflow transition request transitions model for specified request id.
transitions_model = workflow.get_workflow_transition_request_transitions_model(request_id)
print(f"Transitions_model : {transitions_model}")

result = workflow.update_workflow_transition_request(request_id, transitions_model)
print(result)

result = workflow.approve_workflow_transition_request(request_id)
print(result)
```
