# Notebook voor verschillende handelingen op topics in askdelphi
Voordat we kunnen starten moeten we eerst zorgen dat de authenticatie op orde is. Hiervoor hebben we een ASKDELPHI_PORTAL_CODE nodig.

```python
import os
import dotenv

dotenv.load_dotenv()

os.getenv("ASKDELPHI_PORTAL_CODE")
```

```python
from askdelphi_client import AskDelphiClient

client = AskDelphiClient()

client.authenticate()
```

Voor het verwijderen van een topic hebben we het topicVersionId nodig, deze krijgen we door onderstaande code te runnen.

```python
topicId = "1bcbcc7e-cbf8-4864-9b0f-c015b9ba9980"
endpoint = f"/v1/tenant/{{tenantId}}/project/{{projectId}}/acl/{{aclEntryId}}/topic/{topicId}/workflowstate"

result = client._request("POST", endpoint,
                json_data={
                    "action": 1
                })

topicVersionId = result["topicVersionId"]
```

Vervolgens kunnen we de volgende code runnen, om het topic daadwerkelijk te verwijderen.

```python
endpoint = f"v3/tenant/{{tenantId}}/project/{{projectId}}/acl/{{aclEntryId}}/topic/{topicId}/topicVersion/{topicVersionId}"

client._request("DELETE", endpoint,
                json_data={
                    "workflowActions": {}
                })
```
